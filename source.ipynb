```python
from flask import Flask, request, render_template, send_from_directory
from pyngrok import ngrok
import tensorflow as tf
import numpy as np
from PIL import Image
import io
import os

app = Flask(__name__)

# Load your model
model = tf.keras.models.load_model("mnist_digit_recognition_model.h5")
UPLOAD_FOLDER = 'uploads'
os.makedirs(UPLOAD_FOLDER, exist_ok=True)
@app.route("/", methods=["GET", "POST"])
def index():
    if request.method == "POST":
        file = request.files.get("file")
        file_path = os.path.join(UPLOAD_FOLDER, file.filename)
        file.save(file_path)
        if file:
            # Load the image from the uploaded file
            image = Image.open(file)

            # Convert to grayscale
            image = image.convert("L")

            # Resize to 28x28 pixels
            image = image.resize((28, 28))

            # Convert to numpy array
            image_array = np.array(image)

            # Invert colors if needed (white background assumption)
            if image_array.mean() > 127:
                image_array = np.invert(image_array)

            # Normalize pixel values
            image_array = image_array / 255.0

            # Reshape for model input
            image_array = image_array.reshape(1, 28, 28)

            # Get the prediction
            prediction = model.predict(image_array)
            predicted_digit = np.argmax(prediction)

            return f'''
            <html>
            <head>
              <style>
                body {{
                  background-image: url("https://media.istockphoto.com/id/1281736074/vector/abstract-geometric-pattern-background.jpg?s=612x612&w=0&k=20&c=rTHAuJPGgbaGZCDh6PAoVkJr3mTvn5WppwDHAPfNgoU=");
                  background-size: cover;
                  background-position: center;
                  background-repeat: no-repeat;
                  height: 100vh;
                  margin: 0;
                  font-family: Arial, sans-serif;
                  color: black;
                  display: flex;
                  justify-content: center;
                  align-items: center;
               }}

        </style>
      </head>
      <body>
          <div class="result">
              <h1>Predicted Digit: {predicted_digit}</h1>
               <img src="/uploads/{file.filename}" alt="Uploaded Image">
          </div>
      </body>
      </html>
      '''


    return '''
    <html>
    <head>
        <style>
            body {
                background-image: url("https://media.istockphoto.com/id/1281736074/vector/abstract-geometric-pattern-background.jpg?s=612x612&w=0&k=20&c=rTHAuJPGgbaGZCDh6PAoVkJr3mTvn5WppwDHAPfNgoU=");
                background-size: cover;
                background-position: center;
                background-repeat: no-repeat;
                height: 100vh;
                margin: 0;
                font-family: Arial, sans-serif;
                color: black;
            }
            .content {
                text-align: center;
                padding-top: 20vh;
            }
            form {
                display: inline-block;
                margin-top: 20px;
            }
        </style>
    </head>
    <body>
        <div class="content">
            <h1>Upload an Image</h1>
            <form method="POST" enctype="multipart/form-data">
                <input type="file" name="file">
                <input type="submit" value="Predict">
            </form>

        </div>
    </body>
    </html>
    '''
@app.route('/uploads/<filename>')
def send_file(filename):
    return send_from_directory(UPLOAD_FOLDER, filename)

# Start ngrok tunnel to access the Flask app externally
public_url = ngrok.connect(5000)
print(" * ngrok URL:", public_url)

# Run Flask app
app.run(port=5000)
```
